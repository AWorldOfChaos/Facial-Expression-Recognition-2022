```python
from sklearn.linear_model import LinearRegression 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')

x_train = np.array(train['x'])
y_train = np.array(train['y'])

x_test = np.array(test['x'])
y_test = np.array(test['y'])


```

```python
w = 0
b = 0

alpha = 0.00001

def cost(slope, intercept, x, y):
    sum = 0
    for i in range(len(x)):
        sum += (slope*x[i]+intercept-y[i])**2
    return sum/ (2*len(x))

def dcdw(slope, intercept, x, y):
    sum = 0
    for i in range(len(x)):
        sum += (slope*x[i]+intercept-y[i])*x[i]
    return sum/(len(x))

def dcdb(slope, intercept, x, y):
    sum = 0
    for i in range(len(x)):
        sum += (slope*x[i]+intercept-y[i])
    return sum/(len(x))

costarr=[]

for i in range(300):
    costarr.append(cost(w,b,x_train, y_train))
    tempw = w - alpha*dcdw(w,b,x_train, y_train)
    tempb = b - alpha*dcdb(w,b,x_train, y_train)
    w = tempw
    b = tempb

x = np.arange(0,300)
plt.plot(x,costarr)
```

```python
plt.figure(figsize=(4,3))
ax = plt.axes()

ax.scatter(x_test, y_test)
y = w*x_test + b
ax.plot(x_test, y, color="red")

print(w,b)
plt.show()
```
